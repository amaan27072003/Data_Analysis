# Data Analysis On CSV File.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# The file name of your dataset
file_name = 'Auto Sales data.csv'

# Read the CSV file into a DataFrame
try:
    df = pd.read_csv(file_name)
except FileNotFoundError:
    print(f"Error: The file '{file_name}' was not found.")
    exit()

# Display the first 5 rows of the DataFrame
print("\nFirst 5 rows of the DataFrame:")
print(df.head())

# Display the column information
print("\nDataFrame Info:")
print(df.info())

# Perform the data analysis using groupby() and sum()
# Group by 'PRODUCTLINE' and sum the 'SALES'
sales_by_productline = df.groupby('PRODUCTLINE')['SALES'].sum().reset_index()

# Sort the data for better visualization
sales_by_productline = sales_by_productline.sort_values(by='SALES', ascending=False)

print("\nTotal Sales by Product Line:")
print(sales_by_productline)
```

```python
# Create a bar chart
plt.figure(figsize=(12, 7))
plt.bar(sales_by_productline['PRODUCTLINE'], sales_by_productline['SALES'], color='darkblue')
plt.xlabel('Product Line')
plt.ylabel('Total Sales')
plt.title('Total Sales by Product Line')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Save the plot to a file
plot_file = 'sales_by_productline_bar_chart.png'
plt.savefig(plot_file)
print(f"\nPlot saved as '{plot_file}'")
```
